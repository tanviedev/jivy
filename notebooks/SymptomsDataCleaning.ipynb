```python
import pandas as pd

data = [
    # Cardiac
    ["collapsed on ground", "cardiac", "yes", 10],
    ["unconscious", "cardiac", "yes", 10],
    ["not breathing properly", "cardiac", "yes", 10],
    ["gasping for air", "cardiac", "yes", 9],
    ["holding chest and collapsing", "cardiac", "yes", 9],
    ["sweating heavily and pale", "cardiac", "yes", 8],
    ["bluish lips or face", "cardiac", "yes", 9],
    ["confusion or dizziness", "cardiac", "yes", 8],
    ["rapid heartbeat observed", "cardiac", "yes", 8],
    ["nausea or vomiting with chest pain", "cardiac", "yes", 8],

    # Obstetric / Pregnancy
    ["heavy vaginal bleeding during pregnancy", "obstetric", "yes", 10],
    ["pregnant woman collapsed", "obstetric", "yes", 10],
    ["severe abdominal pain during pregnancy", "obstetric", "yes", 9],
    ["baby not moving", "obstetric", "no", 9],
    ["seizure observed in pregnant woman", "obstetric", "yes", 10],
    ["extreme swelling observed", "obstetric", "yes", 8],
    ["high fever in pregnant woman", "obstetric", "yes", 8],
    ["unconscious after labor complication", "obstetric", "yes", 10],
    ["excessive vomiting", "obstetric", "yes", 8],

    # Trauma / Road Accidents
    ["road accident with bleeding", "trauma", "yes", 10],
    ["unconscious after accident", "trauma", "yes", 10],
    ["head injury with bleeding", "trauma", "yes", 9],
    ["open wound with bone visible", "trauma", "yes", 9],
    ["multiple injuries", "trauma", "yes", 9],
    ["minor accident with pain", "trauma", "yes", 6],
    ["limb deformity / fracture observed", "trauma", "yes", 8],
    ["severe bleeding from limb", "trauma", "yes", 9],
    ["person trapped under object", "trauma", "yes", 10],
    ["dazed or confused after accident", "trauma", "yes", 8],

    # General Emergency / Other
    ["person not responding to calls", "general", "yes", 10],
    ["bluish lips or face", "general", "yes", 10],
    ["vomiting continuously", "general", "yes", 8],
    ["high fever with confusion", "general", "yes", 8],
    ["seizure observed", "general", "yes", 10],
    ["difficulty speaking", "general", "yes", 9],
    ["slurred speech or facial droop", "general", "yes", 9],
    ["sudden severe headache", "general", "yes", 8],
    ["sudden collapse with convulsions", "general", "yes", 10],
    ["unresponsive after fall", "general", "yes", 9],
]

# Create DataFrame
symptoms_df = pd.DataFrame(data, columns=["symptom", "emergency_type", "is_visible", "risk_score"])

# Convert 'is_visible' to binary for ML
symptoms_df["is_visible"] = symptoms_df["is_visible"].apply(lambda x: 1 if x=="yes" else 0)

# Save CSV
symptoms_df.to_csv("ml_ready_symptoms_extended.csv", index=False)

symptoms_df.head(20)  # Preview first 20
```

```python
def compute_combined_risk(symptoms_list, df=symptoms_df):
    """
    symptoms_list: list of symptom names observed
    Returns: combined risk score (1-10)
    """
    matched = df[df["symptom"].isin(symptoms_list)]

    if matched.empty:
        return 1  # default low risk if unknown

    # Weighted average considering visibility
    combined_score = (matched["risk_score"] * (1 + matched["is_visible"])).sum() / (len(matched) + matched["is_visible"].sum())

    return min(round(combined_score,1), 10)

# Example
observed = ["chest pain","bluish lips or face"]
print("Combined Risk Score:", compute_combined_risk(observed))
```
