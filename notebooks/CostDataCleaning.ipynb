```python
def get_cost_estimate(emergency_type, risk_score, df=costs_df):
    match = df[
        (df["emergency_type"] == emergency_type) &
        (df["risk_min"] <= risk_score) &
        (df["risk_max"] >= risk_score)
    ]
    return match
```

Cost estimates are indicative ranges based on publicly known healthcare patterns and are intended only for emergency navigation and decision support. They do not represent hospital billing or guarantees.
